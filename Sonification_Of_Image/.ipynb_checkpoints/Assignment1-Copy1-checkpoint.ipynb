```python
%pylab inline
from __future__ import print_function
from __future__ import division
```

```python
from scipy.io import wavfile
```

```python
img = imread('image1.png')
```

```python
imshow(img)
```

```python
sig = img[120,:, 1]
plot(sig)
sig.dtype
```

```python
sigout = list((sig - 100) * 200)*100
plot(sigout)
```

```python
wavfile.write('sample2.wav', 8000, array(sigout, dtype=int16))
```

```python
audioout = array(((img[:,:,0].astype(int16)- 100)).flat)
plot(audioout)
wavfile.write('outsig2.wav', 44100, audioout)
```

```python
audioout = array(((img[:,:,0].astype(int16)- 100) * 200).flat)
wavfile.write('outsig2.wav', 8000, audioout)
print (audioout)
```
