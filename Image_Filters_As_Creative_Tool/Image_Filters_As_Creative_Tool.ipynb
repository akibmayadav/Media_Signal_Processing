## EXPLORING IMAGE FILTERS AS A CREATIVE PROCESSING TOOL


```python
%pylab inline
rcParams['figure.figsize'] = (10, 4)
from __future__ import print_function
from __future__ import division
from scipy.signal import convolve2d
```

```python
img1 = imread('potrait1.jpg')
imshow(img1)
```

```python
mat = zeros((100,100))
for i in range (0,100):
    for j in range (0,100):
        dis = ((i-50)*(i-50))+((j-50)*(j-50))
        if ( dis < 90):
            mat[i,j]=1
imshow(mat, cmap=cm.gray, interpolation='nearest')
```

```python
c = convolve2d(mat, img1[:,:,1])
# kernel is the analog for 
imshow(c, cmap= cm.gray, interpolation='nearest')
colorbar()
```
