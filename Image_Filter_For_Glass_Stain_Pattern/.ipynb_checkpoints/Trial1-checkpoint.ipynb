## Python wave module

```python
import wave
```

```python
fp = wave.open("ambika_mus.wav", 'r')
print(fp)
#If file is a string, open the file by that name. # r implies read only
```

```python
print(fp.getframerate(), fp.getnchannels(), fp.getnframes(), fp.getsampwidth(), fp.getcomptype(), fp.getcompname())
# gives different parameters of the audio file.
```

```python
fp.getparams()
# gives all the parameters
```

```python
bytes = fp.readframes(8)
print(bytes)
```

```python
for byte in bytes:
    print(ord(byte)) 
```

## Image reading

```python
img = imread('fruits.png')
#to read image 
```

```python
imshow(img)
#to display image
```

```python
img.dtype
#data type of image
```

```python
img.shape
# shape of the image ,last value tells number of pixels
```

```python
subplot(131)
imshow(img[:,:,0])
# display the amount of red value
colorbar()
subplot(132)
imshow(img[:,:,1])
# display the amount of green value
colorbar()
subplot(133)
imshow(img[:,:,2])
# display the amount of blue value
colorbar()

gcf().set_figwidth(14)
```

```python
subplot(131)
imshow(img[:,:-30,0], cmap=cm.gray)
colorbar()
subplot(132)
imshow(img[:,:-30,1], cmap=cm.gray)
colorbar()
subplot(133)
imshow(img[:,:-30,2], cmap=cm.gray)
colorbar()

gcf().set_figwidth(14)
# showing R,G B content in terms of grayscale.
# -30 is RX (something called 2D spatial refrencing)
```

```python
imshow(img[:,:,0] + img[:,:,1], cmap=cm.gray)
colorbar()
# adding red and green components and displaying using grayscale
```

```python
imshow(img[:,:,0]/2 + img[:,:,1]/2, cmap=cm.gray)
colorbar()
#averaging out the red and green values
```

```python
imshow(img[:,:,0].astype(int16) + img[:,:,1].astype(int16), cmap=cm.gray)
colorbar()
#convert the rgb values to 16 bit integers
```

```python
imshow(sum(img, axis=2), cmap=cm.gray)
colorbar()
#what is the axis?
```

```python
red_only = img.copy()
red_only[:,:,1] = 0
red_only[:,:,2] = 0
imshow(red_only)
#show red component in red color itself
```

```python
red_only = img.copy()
red_only[:,:,0] = 0
red_only[:,:,2] = 0
imshow(red_only)
```
