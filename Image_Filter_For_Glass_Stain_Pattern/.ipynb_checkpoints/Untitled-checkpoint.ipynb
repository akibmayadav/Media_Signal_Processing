```python
%pylab inline
from __future__ import print_function
from __future__ import division
from scipy.io import wavfile
from PIL import Image
from scipy.spatial import Voronoi, voronoi_plot_2d
```

```python
img1 = imread('img1.png')
```

```python
imshow(img1)
shape(img1)
```

```python
# STEP 1.1 Resizing the image to be divided in the right portion
# STEP 1.2 Making an array of x and y co-ordinates in the image 
def segmentimage(img,blocksize):
    x_s = img.shape[0]
    y_s = img.shape[1]
    x_d = x_s%blocksize
    y_d = y_s%blocksize
    image = img1[0:x_s-x_d+1,0:y_s-y_d+1]
    x_array = []
    y_array = []
    x = image.shape[0]
    y = image.shape[1]
    n_x = int (x/blocksize)
    n_y = int (y/blocksize)
    for i in range(0,n_x+1):
        x_array.append(i*blocksize)
    for i in range(0,n_y+1):
        y_array.append(i*blocksize)
    return x_array,y_array
```

```python
def randomxyint(x,y):
    vornoimat=[]
    for i in range(0,len(x)-1):
        for j in range (0,len(y)-1):
            x_rand=random.randint(x[i],x[i+1])
            y_rand=random.randint(y[j],y[j+1])
            vornoimat.append([x_rand,y_rand])
    return vornoimat

a = randomxyint(segmentimage(img1,6)[0],segmentimage(img1,6)[1])


            
```

```python
# Appending all the vornoi points in matrix a . So now a contains all the points corresponding to the vornoi mesh.
region = []
for i in range(0,len(a)):
    region.append(a[i])
```

```python
vor = Voronoi(randomxyint(segmentimage(img1,6)[0],segmentimage(img1,6)[1]))
voronoi_plot_2d(vor)
plt.show()
```

```python
# TRIAL OF DISTANCE CALCULATION
for i in range(1,img1.shape[0]):
    for j in range(1,img1.shape[1]):
        dis =[]
        for l in range (len(a)):
            distance = ((i-a[l][0])*(i-a[l][0])) + ((j-a[l][1])*(j-a[l][1])^2)
            dis.append(distance)
        region[dis.index(min(dis))].append([i,j])
        
```
